In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
#Your Code Here

%matplotlib notebook

import matplotlib.pyplot as plt
import scipy.stats as stats

# Data
df_T = df.transpose()

# Calculate the unbiased standard error of the mean
n = len(df_T)
mean_values = np.mean(df_T, axis=0)
sem_values = np.std(df_T, axis=0, ddof=1) / np.sqrt(n)
# Calculate the 95% confidence interval
confidence = 0.95
error = stats.t.ppf((1 + confidence) / 2, n - 1) * sem_values
confidence_interval = (mean_values - error, mean_values + error)

# Create a figure and axis
fig, ax = plt.subplots()
# Plot the bar chart with error bars
ax.bar(df_T.columns, mean_values, yerr=error, capsize=4)
# Display the plot
plt.show()
```

```python
year =  [str(x) for x in df.index]
# Calculate the mean and standard deviation for each year
means = np.mean(df, axis=1)
stds = np.std(df, axis=1, ddof=1)
# Calculate the sample size and degrees of freedom
n = len(df_T)
dof = n - 1
# Calculate the 95% confidence interval for each year
confidence_intervals = [stats.t.interval(0.95, dof, loc=mean, scale=std / np.sqrt(n)) for mean, std in zip(means, stds)]
lower_bounds = [interval[0] for interval in confidence_intervals]
upper_bounds = [interval[1] for interval in confidence_intervals]
# User input for the y-axis value of interest
y_value = float(input("Enter the y-axis value of interest: "))
# Create a figure and axis
fig, ax = plt.subplots()
# Plot the bar chart
ax.bar(year, means, yerr=[means - lower_bounds, upper_bounds - means], capsize=4)
# Plot the bar chart and color bars based on the y-axis value of interest
bar_color = []
for i in range(len(year)):
    bar_color.append('white')  # Default color
    if y_value < lower_bounds[i]:
        bar_color[i] = 'red'  # Definitely above the value
    elif y_value > upper_bounds[i]:
        bar_color[i] = 'blue'  # Definitely below the value
# Plot the bar chart
ax.bar(year, means, capsize=4, color=bar_color, edgecolor = 'black')
# Draw a horizontal line at the y-axis value of interest
ax.axhline(y_value, color='black', linestyle='--')
# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Mean Values')
ax.set_title('Bar Chart with 95% Confidence Interval')
# Display the plot
plt.show()
```
